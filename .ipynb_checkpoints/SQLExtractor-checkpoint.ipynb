```python
from dotenv import load_dotenv
import os
from typing import List, Dict, Any, Optional, Union

load_dotenv()

GIGACHAT_CREDENTIALS = os.environ.get("GIGACHAT_CREDENTIALS")

import os
from Classes.model_classes import LangChainSQLLineageExtractor
```

```python
extractor = LangChainSQLLineageExtractor(
    credentials=GIGACHAT_CREDENTIALS,   # or set env var GIGACHAT_API_KEY
    model="GigaChat",
    verify_ssl_certs=False,
    temperature=0.1
)

sql = """
    CREATE VIEW sales.customer_summary AS
    SELECT c.customer_id, c.customer_name, COUNT(o.order_id) as total_orders
    FROM sales.customers c
    LEFT JOIN sales.orders o ON c.customer_id = o.customer_id
    GROUP BY c.customer_id, c.customer_name
    """
lineage = await extractor.extract_lineage(sql)
print(lineage)
```
